### Read CSV back in to further clean data

```python
atl_restaurants_df = pd.read_csv("../yelp_atl_restaurants_csv.csv")
```

```python
categories_col_list = []
for index, row in atl_restaurants_df.iterrows():
    categories_col_list.append(row["categories"])
```

```python
categories_dict = {}

for categories in categories_col_list:
    cat_list = categories.split(", ")
    for item in cat_list:
        if categories_dict.get(item, False):
            categories_dict[item] += 1
        else:
            categories_dict[item] = 1
            
remove_list = ["Restaurants","Food","Bars","Nightlife","Event Planning & Services","Coffee & Tea",
               "Caterers","Desserts","Cocktail Bars","Lounges","Wine Bars","Wine & Spirits","Beer",
               "Sports Bars", "Ice Cream & Frozen Yogurt", "Venues & Event Spaces", "Hookah Bars",
               "Beer Bar", "Shopping", "Food Delivery Services", "Food Stands", "Gluten-Free", "Arts & Entertainment",
               'Juice Bars & Smoothies', "Grocery", "Food Court", "Automotive", "Convenience Stores",
               "Dance Clubs", "Breweries", "Local Flavor", "Bubble Tea", "Dive Bars"]

for item in remove_list:
    categories_dict.pop(item)
            
new_categories = []

for key, value in categories_dict.items():
    if value > 60:
        new_categories.append(key)
        
new_categories
```

```python
categories_df = pd.DataFrame({"Category": new_categories})
categories_df.to_csv("../categories.csv")
categories_df.to_json("../categories.json")
```

```python
search_string = ""

for index, cat in enumerate(new_categories):
    if index == 0:
        search_string = cat
    else:
        search_string += f"|{cat}"
    
atl_restaurants_df["include"] = atl_restaurants_df["categories"].str.contains(search_string, regex=True)
atl_restaurants_df = atl_restaurants_df.loc[atl_restaurants_df["include"] == True, :]
del atl_restaurants_df["include"]
atl_restaurants_df
```

```python
atl_restaurants_df.to_json("../yelp_atl_restaurants.json")
```

```python
atl_restaurants_df.to_csv("../yelp_atl_restaurants_csv.csv")
```
