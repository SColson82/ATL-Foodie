```python
import pandas as pd
```

```python
restaurant_df=pd.read_csv("../yelp_atl_restaurants_csv.csv").drop(['Unnamed: 0'], axis=1).drop(['Unnamed: 0.1'], axis=1)
restaurant_df.head()
```

```python
import sqlite3
conn = sqlite3.connect('db.sqlite')
```

```python
restaurant_df.to_sql('restaurant_db', conn, if_exists='replace', index=False)
```

```python
pd.read_sql('select * from restaurant_db', conn)
```

```python
category_df=pd.read_csv("../categories.csv").drop(['Unnamed: 0'], axis=1)
category_df
```

```python
category_df.to_sql('category_db', conn, if_exists='replace', index=False)
```

```python
pd.read_sql('select * from category_db', conn)
```
