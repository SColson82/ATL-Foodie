```python
import json
import pandas as pd
import numpy as np
```

```python
data_file = open("C:/Users/Sharon/datacode/Class-Projects/ATL-Foodie-heck yea/SourceData/yelp_academic_dataset_review.json", encoding='utf8') 
data = []
for line in data_file:
    data.append(json.loads(line))
df = pd.DataFrame(data)
data_file.close()
```

```python
atl_restaurant_df=pd.read_csv("C:/Users/Sharon/datacode/Class-Projects/ATL-Foodie-heck yea/Resources/yelp_atl_restaurants_csv.csv")
atl_restaurant_df.head()
```

```python
atl_restaurant_review_df=pd.merge(atl_restaurant_df,df, on="business_id", how ="left")
atl_restaurant_review_df
```

```python
clean_atl_restaurant_review_df = atl_restaurant_review_df.drop(['Unnamed: 0'], axis=1).dropna(how='any')
clean_atl_restaurant_review_df
```

```python
clean_atl_restaurant_review_df.head(2)
```

```python
clean_atl_restaurant_review_df.isnull().sum()
```

```python
clean_atl_restaurant_review_df.drop(clean_atl_restaurant_review_df[clean_atl_restaurant_review_df['is_open']==0].index,inplace=True)
clean_atl_restaurant_review_df
```

```python
clean_atl_restaurant_review_df.describe()
```

```python
clean_atl_restaurant_review_df2=clean_atl_restaurant_review_df.rename(columns={'stars_x':'stars(avg)','stars_y':'stars(review)'}).drop(['is_open'],axis=1)
clean_atl_restaurant_review_df2
```

```python
final_atl_restaurant_review_df=clean_atl_restaurant_review_df2.sample(n=40000)
final_atl_restaurant_review_df
```

```python
final_atl_restaurant_review_df.to_json("yelp_atl_restaurant_reviews_json.json")
```
